```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://quotes.toscrape.com/"

response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")

print(soup.title)
```

```python
quotes = []
authors = []
tags_list = []

quote_containers = soup.find_all("div", class_="quote")

for container in quote_containers:

    quote = container.find("span", class_="text").text
    author = container.find("small", class_="author").text
    tags = container.find_all("a", class_="tag")

    tag_text = [tag.text for tag in tags]

    quotes.append(quote)
    authors.append(author)
    tags_list.append(", ".join(tag_text))
```

```python
data = pd.DataFrame({
    "Quote": quotes,
    "Author": authors,
    "Tags": tags_list
})

data
```

```python
data.to_csv("quotes_dataset.csv", index=False)

print("CSV file created successfully.")
```
