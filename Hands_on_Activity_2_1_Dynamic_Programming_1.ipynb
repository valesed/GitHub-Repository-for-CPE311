0/1 Knapsack Problem

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

```python
#type your code here for recursion programming solution

bills = [1000, 500, 100, 50, 10, 5, 1]

def change(amount, index, result):
    if amount == 0:
        return True
    if amount < 0 or index == len(bills):
        return False

    bill = bills[index]
    max_use = amount // bill

    for i in range(max_use, -1, -1):
        result[bill] = i
        if change(amount - bill * i, index + 1, result):
            return True

    result[bill] = 0
    return False

bill = int(input("Enter bill amount: "))
paid = int(input("Enter amount paid: "))

change_amount = paid - bill

if change_amount < 0:
    print("Not enough payment.")
else:
    result = {}
    if change(change_amount, 0, result):
        print("Change:", change_amount)
        print("Change breakdown:")
        for b in sorted(result.keys(), reverse=True):
            if result[b] > 0:
                print("₱", b, ":", result[b])
    else:
        print("Exact change not possible with available bills/coins.")

```

```python
#type your code here for dynamic programming solution

bills = [1, 5, 10, 50, 100, 500, 1000]

bill = int(input("Enter bill amount: "))
paid = int(input("Enter amount paid: "))

change = paid - bill

if change < 0:
    print("Not enough payment.")
else:
    dp = [float('inf')] * (change + 1)
    dp[0] = 0
    last_bill = [0] * (change + 1)

    for i in range(1, change + 1):
        for b in bills:
            if i - b >= 0 and dp[i - b] + 1 < dp[i]:
                dp[i] = dp[i - b] + 1
                last_bill[i] = b

    if dp[change] == float('inf'):
        print("Exact change not possible with available bills/coins.")
    else:
        print("Change:", change)
        result = {}
        amount = change
        while amount > 0:
            b = last_bill[amount]
            result[b] = result.get(b, 0) + 1
            amount -= b

        print("Change breakdown:")
        for b in sorted(result.keys(), reverse=True):
            print("₱", b, ":", result[b])
```
