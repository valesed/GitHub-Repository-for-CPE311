```python
import pandas as pd

data = pd.read_csv("/content/2019_Yellow_Taxi_Trip_Data.csv")

#Create a DataFrame by reading in the 2019_Yellow_Taxi_Trip_Data.csv file.
data.head()
```

```python
#Examine the first 5 rows.
data.tail(5)
```

```python
#Find the dimensions (number of rows and number of columns) in the data.
data.info()
```

```python

selected_features = ["fare_amount", "tip_amount", "tolls_amount", "total_amount"]

# Calculate summary statistics for the selected features
display(data[selected_features].describe())
```

```python
#Isolate the fare_amount, tip_amount, tolls_amount, and total_amount for the longest trip by distance (trip_distance).
longest_trip = data.sort_values(by='trip_distance', ascending=False).head(5)
display(longest_trip[selected_features])
```
