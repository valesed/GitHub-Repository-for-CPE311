# Hands-on Activity 2.1: The Tower of Hanoi Problem

```python
def print_pegs(pegs):
    print(f"A: {pegs['A']}")
    print(f"B: {pegs['B']}")
    print(f"C: {pegs['C']}")
    print("_" * 25)

def tower(n, start, aux, end, pegs):
    if n == 1:
        disk = pegs[start].pop()
        pegs[end].append(disk)
        print(f"Move disk {disk} from {start} to {end}")
        print_pegs(pegs)
        return

    tower(n - 1, start, end, aux, pegs)
    disk = pegs[start].pop()
    pegs[end].append(disk)
    print(f"Move disk {disk} from {start} to {end}")
    print_pegs(pegs)
    tower(n - 1, aux, start, end, pegs)

pegs = {
    "A": [4, 3, 2, 1],
    "B": [],
    "C": []
}
print("Initial State:")
print_pegs(pegs)
tower(4, "A", "B", "C", pegs)
```
