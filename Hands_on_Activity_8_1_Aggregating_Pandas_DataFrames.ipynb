```python
import pandas as pd
import numpy as np
```

```python
earthquakes = pd.read_csv("earthquakes.csv")
faang = pd.read_csv("faang.csv")

earthquakes.head()
faang.head()
```

```python
japan_quakes = earthquakes[
    (earthquakes["place"].str.contains("Japan", na=False)) &
    (earthquakes["magType"] == "mb") &
    (earthquakes["mag"] >= 4.9)
]

japan_quakes
```

```python
ml_quakes = earthquakes[earthquakes["magType"] == "ml"]

bins = np.arange(0, ml_quakes["mag"].max() + 1)

ml_quakes["mag_bin"] = pd.cut(ml_quakes["mag"], bins)

bin_counts = ml_quakes.groupby("mag_bin").size()

bin_counts
```

```python
faang["date"] = pd.to_datetime(faang["date"])
faang.set_index("date", inplace=True)

monthly = faang.groupby("ticker").resample("M").agg({
    "open": "mean",
    "high": "max",
    "low": "min",
    "close": "mean",
    "volume": "sum"
})

monthly
```

```python
crosstab = pd.crosstab(
    earthquakes["tsunami"],
    earthquakes["magType"],
    values=earthquakes["mag"],
    aggfunc="max"
)

crosstab
```

```python
rolling_60 = faang.groupby("ticker").rolling("60D").agg({
    "open": "mean",
    "high": "max",
    "low": "min",
    "close": "mean",
    "volume": "sum"
})

rolling_60
```

```python
pivot_table = pd.pivot_table(
    faang,
    values=["open", "high", "low", "close", "volume"],
    index="ticker",
    aggfunc="mean"
)

pivot_table
```

```python
nflx = faang[faang["ticker"] == "NFLX"]

z_scores = nflx.select_dtypes(include=np.number).apply(
    lambda x: (x - x.mean()) / x.std()
)

z_scores.head()
```

```python
events = pd.DataFrame({
    "ticker": ["FB", "FB", "FB"],
    "date": ["2018-07-25", "2018-03-19", "2018-03-20"],
    "event": [
        "Disappointing user growth announced after close.",
        "Cambridge Analytica story",
        "FTC investigation"
    ]
})

events["date"] = pd.to_datetime(events["date"])
events.set_index(["date", "ticker"], inplace=True)

faang_reset = faang.reset_index().set_index(["date", "ticker"])

merged = faang_reset.merge(events, how="outer", left_index=True, right_index=True)

merged
```

```python
indexed = faang.groupby("ticker").transform(lambda x: x / x.iloc[0])

indexed.head()
```
