```python
import pandas as pd
import numpy as np

meteorites = pd.read_csv("/content/meteorite-landings.csv")
meteorites['year'] = pd.to_numeric(meteorites['year'], errors='coerce')
meteorites_filtered = meteorites[(meteorites['year'] >= 2005) & (meteorites['year'] <= 2009)]

pivot_table = meteorites_filtered.pivot_table(
    index='year',
    columns='fall',
    values='mass',
    aggfunc=[len, lambda x: np.percentile(x.dropna(), 95)]
)
pivot_table

summary_stats = meteorites_filtered.groupby('fall')['mass'].describe()
summary_stats
```

```python
taxi = pd.read_csv("/content/2019_Yellow_Taxi_Trip_Data.csv", parse_dates=['tpep_dropoff_datetime'])
taxi.set_index('tpep_dropoff_datetime', inplace=True)
hourly_summary = taxi.resample('H')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum()
top_5_tips = hourly_summary.nlargest(5, 'tip_amount')
top_5_tips
```

The Meteorite Landings dataset from Kaggle contains worldwide meteorite records, including mass, year, and whether the meteorite was observed falling or simply found, allowing analysis of meteorite trends over time. The 2019 Yellow Taxi Trip dataset from GitHub contains NYC taxi trip records, including pickup/dropoff times, trip distance, fares, tips, and tolls, enabling analysis of hourly taxi activity and revenue patterns.
