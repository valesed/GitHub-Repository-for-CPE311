```python
import pandas as pd

df = pd.read_csv('life expentancy rate.csv')
print(df.head())
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean(numeric_only=True))
```

### 3. Prepare the data by applying appropriate data preprocessing techniques.

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
```

### 4. Analyze the data using descriptive analysis.

```python
display(df_numeric.describe())
```

### 4. Perform correlation analysis.

```python
correlation = df_numeric.corr()
display(correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Correlation Matrix of Life Expectancy Data")
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.tight_layout()
plt.show()
```

```python
year_columns = [col for col in df_numeric.columns if col.startswith(('20'))]
years = sorted(list(set([col[:4] for col in year_columns])))

avg_values = pd.Series(dtype=float)
for year in years:
    cols_for_year = [col for col in year_columns if col.startswith(year)]
    if not df_numeric[cols_for_year].empty:
        avg_values[year] = df_numeric[cols_for_year].mean().mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_values.index, avg_values.values, marker='o')
plt.xticks(rotation=45)
plt.title("Average Life Expectancy Trend (ASEAN)")
plt.xlabel("Year")
plt.ylabel("Average Life Expectancy")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(avg_values.index, avg_values.values, color='skyblue')
plt.xticks(rotation=45)
plt.title("Average Life Expectancy Comparison (ASEAN)")
plt.xlabel("Year")
plt.ylabel("Average Life Expectancy")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_melted = df.melt(id_vars=['Countries_and_areas', 'Reference_year(s)', 'Source'],
                    var_name='Year_Gender',
                    value_name='Life_Expectancy')

df_melted['Year'] = df_melted['Year_Gender'].str[:4].astype(int)
df_melted['Gender'] = df_melted['Year_Gender'].str[4:]

df_melted = df_melted.dropna(subset=['Life_Expectancy'])

plt.figure(figsize=(15, 8))
sns.lineplot(data=df_melted, x='Year', y='Life_Expectancy', hue='Countries_and_areas', style='Gender', marker='o')

plt.title('Life Expectancy Trends for Individual ASEAN Countries (2004-2014)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.xticks(sorted(df_melted['Year'].unique()))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_melted_clustered = df_melted_clustered[df_melted_clustered['Gender'] != 'M.1']
print("Removed 'M.1' from Gender column.")
# Re-create df_melted_gender_combined to reflect the change
df_melted_gender_combined = df_melted_clustered.groupby(['Countries_and_areas', 'Year', 'Region'])['Life_Expectancy'].mean().reset_index()
```

### Regional Analysis

```python
regional_mapping = {
    'Brunei Darussalam': 'Maritime Southeast Asia',
    'Cambodia': 'Mainland Southeast Asia',
    'Indonesia': 'Maritime Southeast Asia',
    "Lao's PDR": 'Mainland Southeast Asia',
    'Malaysia': 'Maritime Southeast Asia',
    'Myanmar': 'Mainland Southeast Asia',
    'Philippines': 'Maritime Southeast Asia',
    'Singapore': 'Maritime Southeast Asia',
    'Thailand': 'Mainland Southeast Asia',
    'Viet Nam': 'Mainland Southeast Asia'
}

df_melted_clustered.loc[:, 'Countries_and_areas'] = df_melted_clustered['Countries_and_areas'].str.strip()

df_melted_clustered.loc[:, 'Region'] = df_melted_clustered['Countries_and_areas'].map(regional_mapping)

if df_melted_clustered['Region'].isnull().any():
    print("Warning: Some countries could not be mapped to a region.")
    print(df_melted_clustered[df_melted_clustered['Region'].isnull()]['Countries_and_areas'].unique())

df_melted_gender_combined = df_melted_clustered.groupby(['Countries_and_areas', 'Year', 'Region'])['Life_Expectancy'].mean().reset_index()

unqiue_regions = df_melted_gender_combined['Region'].dropna().unique()

for region in unqiue_regions:
    plt.figure(figsize=(12, 7))
    regional_df = df_melted_gender_combined[df_melted_gender_combined['Region'] == region]

    sns.lineplot(data=regional_df, x='Year', y='Life_Expectancy',
                 hue='Countries_and_areas', marker='o')

    plt.title(f'Average Life Expectancy Trends in {region} (2005-2014)')
    plt.xlabel('Year')
    plt.ylabel('Average Life Expectancy (Years)')
    plt.xticks(sorted(regional_df['Year'].unique()))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
    plt.tight_layout()
    plt.show()
```

### Regional Life Expectancy Trends by Gender

```python
unqiue_regions_gender_separated = df_melted_clustered['Region'].dropna().unique()
genders = df_melted_clustered['Gender'].dropna().unique()

for region in unqiue_regions_gender_separated:
    regional_df_gender_separated = df_melted_clustered[df_melted_clustered['Region'] == region]
    for gender in sorted(genders):
        plt.figure(figsize=(12, 7))
        gender_df = regional_df_gender_separated[regional_df_gender_separated['Gender'] == gender]

        sns.lineplot(data=gender_df, x='Year', y='Life_Expectancy',
                     hue='Countries_and_areas', marker='o')

        plt.title(f'{gender} Life Expectancy Trends in {region} (2004-2014)')
        plt.xlabel('Year')
        plt.ylabel('Life Expectancy (Years)')
        plt.xticks(sorted(gender_df['Year'].unique()))
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
        plt.tight_layout()
        plt.show()
```

```python
unqiue_regions_gender_separated = df_melted_clustered['Region'].dropna().unique()
genders = df_melted_clustered['Gender'].dropna().unique()

for region in unqiue_regions_gender_separated:
    regional_df_gender_separated = df_melted_clustered[df_melted_clustered['Region'] == region]
    for gender in sorted(genders):
        plt.figure(figsize=(12, 7))
        gender_df = regional_df_gender_separated[regional_df_gender_separated['Gender'] == gender]

        sns.lineplot(data=gender_df, x='Year', y='Life_Expectancy',
                     hue='Countries_and_areas', marker='o')

        plt.title(f'{gender} Life Expectancy Trends in {region} (2004-2014)')
        plt.xlabel('Year')
        plt.ylabel('Life Expectancy (Years)')
        plt.xticks(sorted(gender_df['Year'].unique()))
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
        plt.tight_layout()
        plt.show()
```
