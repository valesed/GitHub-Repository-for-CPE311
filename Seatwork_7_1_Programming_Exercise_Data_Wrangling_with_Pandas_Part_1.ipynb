# 1. Extract only the year

```python
import pandas as pd

df = pd.read_csv("/content/meteorite-landings.csv")

df['year'] = df['year'].astype(str).str.slice(0,4)
df

```

# 2. Convert year to numeric

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df
```

# 3. Create new column: meteorite fell before 1970

```python
df['before_1970'] = df['year'] < 1970
df
```

# 4. Set id as index

```python
df = df.set_index('id')
df
```

# 5. Sort index before using loc


```python
df = df.sort_index()
df
```

# 6. Extract rows with id between 10036 and 10040

```python
result = df.loc[10036:10040]
result
```
